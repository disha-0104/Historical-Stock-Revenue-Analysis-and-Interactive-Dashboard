```python
# Question 1: Use yfinance to Extract Tesla Stock Data
import yfinance as yf
import pandas as pd

# Fetch Tesla stock data
tesla = yf.Ticker("TSLA")
tesla_data = tesla.history(period="max")

# Reset the index to display it properly
tesla_data.reset_index(inplace=True)

# Display the first five rows of the tesla_data dataframe
tesla_data.head()
```

```python
# Question 2: Use Webscraping to Extract Tesla Revenue Data
# We are using yfinance for Tesla revenue data
tesla_revenue_data = tesla.financials.loc["Total Revenue"].reset_index()
tesla_revenue_data.columns = ["Date", "Revenue"]

# Display the last five rows of the tesla_revenue dataframe
tesla_revenue_data.tail()
```

```python
# Question 3: Use yfinance to Extract Stock Data for GameStop
# Fetch GameStop stock data
gamestop = yf.Ticker("GME")
gme_data = gamestop.history(period="max")

# Reset the index to display it properly
gme_data.reset_index(inplace=True)

# Display the first five rows of the gme_data dataframe
gme_data.head()
```

```python
# Question 4: Use Webscraping to Extract GameStop Revenue Data
# We are using yfinance for GameStop revenue data
gme_revenue_data = gamestop.financials.loc["Total Revenue"].reset_index()
gme_revenue_data.columns = ["Date", "Revenue"]

# Display the last five rows of the gme_revenue dataframe
gme_revenue_data.tail()
```

```python
# Question 5: Plot Tesla Stock Graph
import matplotlib.pyplot as plt

# Plot Tesla Stock Price Data
plt.figure(figsize=(10, 6))
plt.plot(tesla_data["Date"], tesla_data["Close"], label="Tesla Stock Price", color="blue")
plt.title("Tesla Stock Price Over Time")
plt.xlabel("Date")
plt.ylabel("Stock Price (USD)")
plt.legend(loc="upper left")
plt.grid(True)
plt.show()
```

```python
# Question 6: Plot GameStop Stock Graph
# Plot GameStop Stock Price Data
plt.figure(figsize=(10, 6))
plt.plot(gme_data["Date"], gme_data["Close"], label="GameStop Stock Price", color="red")
plt.title("GameStop Stock Price Over Time")
plt.xlabel("Date")
plt.ylabel("Stock Price (USD)")
plt.legend(loc="upper left")
plt.grid(True)
plt.show()
```
